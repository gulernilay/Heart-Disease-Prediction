# Handle the missing values

```python
## Replacing null values with the mean value
heart_df["BPMeds"].fillna(heart_df["BPMeds"].mean(),inplace=True)
heart_df["totChol"].fillna(heart_df["totChol"].mean(),inplace=True)
heart_df["education"].fillna(heart_df["education"].mean(),inplace=True)
heart_df["cigsPerDay"].fillna(heart_df["cigsPerDay"].mean(),inplace=True)
heart_df["BMI"].fillna(heart_df["BMI"].mean(),inplace=True)
heart_df["heartRate"].fillna(heart_df["heartRate"].mean(),inplace=True)
heart_df["glucose"].fillna(heart_df["glucose"].mean(),inplace=True)
```

```python
## Check null values after imputation
heart_df.isnull().sum()
```

# Data Visualization

```python
#Age Distribution by Heart Disease Status
plt.figure(figsize=(10,6))
sns.histplot(data=heart_df, x="age", hue="TenYearCHD", bins=30, kde=True)
plt.title("Age Distribution by Heart Disease Status")
plt.show()
```

```python
#Boxplot of BMI by Heart Disease Status
plt.figure(figsize=(8,6))
sns.boxplot(x="TenYearCHD", y="BMI", data=heart_df)
plt.title("BMI by Heart Disease Status")
plt.show()
```

```python
#Pairplot of Selected Features
sns.pairplot(heart_df, vars=["age", "BMI", "sysBP", "glucose"], hue="TenYearCHD")
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
##Countplot of Education Levels
plt.figure(figsize=(8,6))
sns.countplot(x="education", hue="TenYearCHD", data=heart_df)
plt.title("Education Level Distribution by Heart Disease Status")
plt.show()
```

```python
#Violin Plot of Cholesterol by Heart Disease Status
plt.figure(figsize=(8,6))
sns.violinplot(x="TenYearCHD", y="totChol", data=heart_df)
plt.title("Cholesterol Distribution by Heart Disease Status")
plt.show()
```
